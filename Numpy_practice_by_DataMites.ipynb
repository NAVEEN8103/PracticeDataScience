```python
# HOW DO YOU FILL A DIAGONAL VALUE WITH THE NUMBER OF A NUMPY ARRAY.
```

```python
my_arr =  np.array([[8,9],[6,7]])
my_arr
```

```python
x = 5
```

```python
# METHOD 1
def filldiag(my_arr, x):
  for i in range(len(my_arr)):
    for j in range(len(my_arr[0])):
      if i ==j :
        my_arr[i][j] = x
  return my_arr
```

```python
filldiag(my_arr,x)
```

```python
# METHOD 2
np.fill_diagonal(my_arr, x)
```

```python
my_arr
```
